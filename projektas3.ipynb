```python
import pandas as pd
```

```python
vp = pd.read_csv('registered_people_n_streets.csv')
vp
```

```python
vp['amzius'] = (2023 - vp['GIMIMO_METAI'])
vp
```

```python
vp.loc[vp['amzius'] <= 30, 'jaunimas'] = 1 
vp.loc[vp['amzius'] > 30, 'jaunimas'] = 0
vp
```

```python
vp['visi'] = 1
vp
```

```python
vp_sum = vp.groupby('SENIUNIJA').sum()[ [ 'visi', 'jaunimas'] ]
vp_sum
```

```python
vp_sum['vyresni_zmones'] = vp_sum['visi'] - vp_sum['jaunimas']
vp_sum
```
